# 3. Variables y Tipos de Datos

En Python, no necesitas declarar el tipo de una variable. Python infiere el tipo automáticamente.



```python
# Entero
x = 10

# Flotante
y = 20.5

# Cadena de texto
nombre = "Juan"

# Booleano
es_estudiante = True

print(x, y, nombre, es_estudiante)
```

# 4. Operadores Básicos

Python soporta operadores aritméticos, de comparación y lógicos.

```python
# Aritméticos
suma = x + y
resta = x - y
multiplicacion = x * y
division = x / y

# Comparación
es_mayor = x > y
es_igual = x == y

# Lógicos
and_logico = (x > 5) and (y < 30)
or_logico = (x > 5) or (y > 30)
not_logico = not (x > 5)

print(suma, resta, multiplicacion, division)
print(es_mayor, es_igual)
print(and_logico, or_logico, not_logico)
```

# 5. Estructuras de Control

Python tiene estructuras de control como if, for, y while.

```python
# Condicional if
if x > 10:
    print("x es mayor que 10")
elif x == 10:
    print("x es igual a 10")
else:
    print("x es menor que 10")
```

```python
# Bucle for
for i in range(5):
    print(i)
```

```python
# Bucle while
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

# 6. Listas

Las listas son colecciones ordenadas y mutables de elementos.

```python
frutas = ["manzana", "banana", "cereza"]
print(frutas[0])  # Acceder al primer elemento
frutas.append("naranja")  # Añadir un elemento
frutas.remove("banana")  # Eliminar un elemento
print(frutas)
```

# 7. Funciones

Las funciones te permiten encapsular código para reutilizarlo.



```python
def saludar(nombre):
    return f"Hola, {nombre}"

mensaje = saludar("Ana")
print(mensaje)
```

# 8. Diccionarios

Los diccionarios son colecciones de pares clave-valor.


```python
persona = {
    "nombre": "Carlos",
    "edad": 25,
    "ciudad": "Madrid"
}

print(persona["nombre"])  # Acceder a un valor
persona["edad"] = 26  # Modificar un valor
persona["ocupacion"] = "Ingeniero"  # Añadir un nuevo par clave-valor
print(persona)
```

# 9. Manejo de Archivos

Python facilita la lectura y escritura de archivos.

```python
# Leer un archivo
with open("archivo.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)

# Escribir un archivo
with open("archivo.txt", "w") as archivo:
    archivo.write("Hola, este es un archivo de texto.")
```

# 10. Módulos y Bibliotecas

Python tiene una gran cantidad de módulos y bibliotecas que puedes importar para extender su funcionalidad.

```python
import math

raiz_cuadrada = math.sqrt(16)
print(raiz_cuadrada)
```

# 11. Manejo de Excepciones

El manejo de excepciones te permite gestionar errores en tu código.

```python
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("No se puede dividir por cero")
finally:
    print("Este bloque siempre se ejecuta")
```

# 12. Clases y Objetos

Python es un lenguaje orientado a objetos. Aquí tienes un ejemplo básico de una clase.

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        return f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años."

persona1 = Persona("Laura", 30)
print(persona1.presentarse())
```
